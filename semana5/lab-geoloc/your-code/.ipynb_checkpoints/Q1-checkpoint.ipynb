El primer paso es seleccionar lo datos desde MongoDB, pasarlos a un dataframe, limpiarlos y guardar las seleccion de datos en un archivo json para volver a llevarlos a MongoDB.

```python
# Tu codigo aqui
from pymongo import MongoClient
import pandas as pd
from pandas.io.json import json_normalize

client=MongoClient()
db=client.companies
```

By office, company name, number of employees, category_code
desarrollo de software, videojuegos, empresas de desarrollo web, e-commerce, mobile, foto, video e incluso de network hosting

```python
companies = list(db.companies.find({'$and': [{'category_code': {'$in': ['web', 'software', 'games_video', 'ecommerce', 
                                                              'mobile', 'photo_video', 'network_hosting']}},
                                            {'number_of_employees': {'$exists': 1}}
                                            ]},
                                {'category_code':1, 'number_of_employees': 1,
                                 'office': 1, 'name':1, '_id':0}))
```
